```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import arff
from io import StringIO
%matplotlib inline
```

# Probablity Distributions



```python
df = pd.read_csv("Iris.csv")

df.head()
```

```python
print('Line plots showing Sepal and Petal data over time')
plt.plot(df['SepalLengthCm'],df['SepalWidthCm'],df['PetalLengthCm'],df['PetalWidthCm'])
plt.show()
```

```python
print('Scatterplot showing the relationship between all Sepal Lengths and Widths and Petal Lentghs and Widths.')
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'],df['PetalLengthCm'],df['PetalWidthCm'])
plt.show()
```

```python
print('Histograms show the distribution of the Sepal Lengths, Widths and Petal Lengths and Widths and any outliers.')
plt.hist(df['SepalLengthCm'])
plt.hist(df['SepalWidthCm'])
plt.hist(df['PetalLengthCm'])
plt.hist(df['PetalWidthCm'])
plt.show()
```

```python
print('QQ plots show how close a variable is to known distribution, and any outliers.')
norm= np.random.normal(0, 1, 150)
norm.sort()

plt.plot(norm, df['SepalLengthCm'].sort_values(), "o")
plt.plot(norm, df['SepalWidthCm'].sort_values(), "o")
plt.plot(norm, df['PetalLengthCm'].sort_values(), "o") 
plt.plot(norm, df['PetalWidthCm'].sort_values(), "o") 
plt.show() 
```

```python
print('Boxplots used to compare groups of Sepals and Petals and to identify differences in variance, as well as outliers.')
plt.boxplot([df['SepalLengthCm'],df['SepalWidthCm'],df['PetalLengthCm'],df['PetalWidthCm']])
plt.show()
```
